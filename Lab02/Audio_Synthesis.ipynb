```python
import numpy as np
from scipy.io.wavfile import write

# CDDA standards
samplerate = 44100  # samples per second
duration = 3      # seconds
frequency = 440   # Hz (A4 note)

# Generate time array
t = np.linspace(0., duration, int(samplerate * duration), endpoint=False)

# Generate sine wave (amplitude normalized to 1 for float representation)
audio = 0.5 * np.sin(2 * np.pi * frequency * t)

# Scale to 16-bit PCM range (-32768 to 32767)
audio_int16 = (audio * 32767).astype(np.int16)

# Write the audio to a WAV file
output_filename = 'sine_wave_cdda.wav'
write(output_filename, samplerate, audio_int16)

print(f"Generated '{output_filename}' (3 seconds, 440 Hz sine wave, CDDA standard).")
```
