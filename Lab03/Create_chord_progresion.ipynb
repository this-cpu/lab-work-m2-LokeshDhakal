```python
from mido import MidiFile, MidiTrack, Message
import pygame

mid = MidiFile()
track = MidiTrack()
mid.tracks.append(track)

# Improved harmonic progression with smooth voice leading + richer chord colors
# Format: ( [notes], duration )
progression = [
    ([60, 64, 67, 71], 960),   # Cmaj7
    ([59, 62, 67, 72], 960),   # Gsus4 â†’ resolves later
    ([57, 60, 64, 69], 960),   # Am7
    ([53, 57, 60, 65], 960),   # Fmaj7
]

for chord, dur in progression:
    # note on
    for n in chord:
        track.append(Message('note_on', note=n, velocity=80, time=0))
    # note off
    for n in chord:
        track.append(Message('note_off', note=n, velocity=64, time=dur))

clip_path = "midi-clips/harmony_piano.mid"
mid.save(clip_path)
pygame.mixer.init()
pygame.mixer.music.load(clip_path)
pygame.mixer.music.play()

print("Playing MIDI...")
while pygame.mixer.music.get_busy():
    pass
```

```python
mid = MidiFile()
track = MidiTrack()
mid.tracks.append(track)

# Set instrument to Nylon Acoustic Guitar (Program 24 in zero-based, Program 25 in GM)
track.append(Message('program_change', program=2, time=0))

# Chords with smooth voice leading
chords = [
    [60, 64, 67, 71],   # Cmaj7
    [59, 62, 67, 72],   # Gsus4
    [57, 60, 64, 69],   # Am7
    [53, 57, 60, 65],   # Fmaj7
]

# Simple picking pattern (arpeggio)
# Pattern offsets represent slight delays for realism
pattern_offsets = [0, 80, 160, 240]

for chord in chords:
    # Play arpeggio
    for i, note in enumerate(chord):
        # note on
        track.append(
            Message("note_on", note=note, velocity=70, time=pattern_offsets[i])
        )
        # note off after short pluck
        track.append(
            Message("note_off", note=note, velocity=50, time=200)
        )

    # Wait before next chord
    track.append(Message("note_on", note=0, velocity=0, time=400))

clip_path = "midi-clips/harmony_guitar.mid"
mid.save(clip_path)
pygame.mixer.init()
pygame.mixer.music.load(clip_path)
pygame.mixer.music.play()

print("Playing MIDI...")
while pygame.mixer.music.get_busy():
    pass
```
